```python


def run_quiz(questions):
    score = 0
    total_questions = len(questions)

    print("\n--- Início do Quiz da Copa do Mundo ---")
    print(f"Você terá que acertar {total_questions} perguntas para ganhar o jogo.\n")

    for i, q in enumerate(questions):
        print(f"Pergunta {i + 1}/{total_questions}: {q['question']}")
        for option_key, option_value in q['options'].items():
            print(f"  {option_key}) {option_value}")

        while True:
            user_answer = input("Sua resposta (A ou B): ").upper()
            if user_answer in ['A', 'B']:
                break
            else:
                print("Resposta inválida. Por favor, digite A ou B.")

        if user_answer == q['correct_answer']:
            print("Correto! Próxima pergunta...\n")
            score += 1
        else:
            print(f"Errado! A resposta correta era {q['correct_answer']}. Fim de jogo.\n")
            return False # Perdeu o jogo

    if score == total_questions:
        print("Parabéns! Você acertou todas as perguntas e ganhou o jogo!\n")
        return True # Ganhou o jogo
    else:
        # Este bloco só seria alcançado se a lógica de 'Errado' acima mudasse.
        # Mas para garantir, caso o jogo termine por algum motivo sem todas as corretas.
        return False

# Sortear 10 perguntas do banco
selected_questions = random.sample(questions_bank, 10)

# Rodar o quiz
run_quiz(selected_questions)
```
