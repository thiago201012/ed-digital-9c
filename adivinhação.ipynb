### Projeto 1: Adivinhe o Número

Este jogo simples pedirá ao usuário para adivinhar um número entre 1 e 10. O programa gerará um número aleatório e verificará se a suposição do usuário está correta.

```python
import random

def guessing_game():
    print("Bem-vindo ao jogo 'Adivinhe o Número'!")
    print("Estou pensando em um número entre 1 e 100. Consegue adivinhá-lo?")

    secret_number = random.randint(1, 100)
    guess = 0

    while guess != secret_number:
        try:
            guess = int(input("Digite seu palpite (um número entre 1 e 100): "))
            if guess < 1 or guess > 100:
                print("Por favor, digite um número entre 1 e 100.")
            elif guess < secret_number:
                print("Muito baixo! Tente novamente.")
            elif guess > secret_number:
                print("Muito alto! Tente novamente.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

    print(f"Parabéns! Você adivinhou o número {secret_number} corretamente!")

guessing_game()
```

### Projeto 2: Jogo de Quiz

Este projeto implementa um quiz com três perguntas. O jogador responde às perguntas e, ao final, recebe uma pontuação e uma mensagem indicando se ganhou o jogo (acertando todas as perguntas).

```python
def quiz_game():
    print("\nBem-vindo ao Jogo de Quiz!")
    print("Responda às seguintes perguntas. Você precisa acertar todas para ganhar!")

    questions = [
        {
            "question": "Qual é a capital da França?",
            "answer": "Paris"
        },
        {
            "question": "Quanto é 2 + 2?",
            "answer": "4"
        },
        {
            "question": "Qual é a cor do céu em um dia claro?",
            "answer": "Azul"
        }
    ]

    score = 0
    for q in questions:
        user_answer = input(q["question"] + " ").strip().capitalize()
        if user_answer == q["answer"].capitalize():
            print("Correto!")
            score += 1
        else:
            print(f"Incorreto. A resposta correta era {q['answer']}.")

    print(f"\nVocê terminou o quiz! Sua pontuação final é: {score} de {len(questions)}.")

    if score == len(questions):
        print("Parabéns! Você acertou todas as perguntas e ganhou o jogo!")
    else:
        print("Infelizmente, você não acertou todas as perguntas. Tente novamente!")

quiz_game()
```
